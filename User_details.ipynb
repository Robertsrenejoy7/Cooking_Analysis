```python
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\Maggy\\OneDrive\\Desktop\\userdetail.csv")
```

```python
df
```

```python
# Step 3: Clean the data
# 3.1 Remove duplicates
df = df.drop_duplicates()

# 3.2 Handle missing values
df = df.dropna(subset=['User ID', 'Email'])  # Drop rows where User ID or Email are missing
df['Total Orders'] = df['Total Orders'].fillna(0)  # Replace missing Total Orders with 0

# 3.3 Standardize text columns (capitalize Location and Favorite Meal)
df['Location'] = df['Location'].str.title()
df['Favorite Meal'] = df['Favorite Meal'].str.capitalize()

# 3.4 Convert date columns to datetime format
df['Registration Date'] = pd.to_datetime(df['Registration Date'], errors='coerce')

# 3.5 Ensure data types are consistent
df['Total Orders'] = df['Total Orders'].astype(int)  # Ensure Total Orders is an integer

# Step 4: Final Cleaned Data Overview
print("\nCleaned Data Overview:")
print(df.head())

# Step 5: Save the cleaned data
df.to_csv("cleaned_data.csv", index=False)
print("\nCleaned data has been saved to 'cleaned_data.csv'.")
```
